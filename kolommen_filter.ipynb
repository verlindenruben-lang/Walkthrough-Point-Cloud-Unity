# Tekstbestand inlezen en eerste 6 kolommen bewaren
Dit notebook leest een tekstbestand in, behoudt enkel de eerste 6 kolommen en verwijdert de laatste 3.

```python
import pandas as pd

# ── Instellingen ──────────────────────────────────────────────────────────────
invoerbestand  = 'invoer.txt'   # Pad naar het tekstbestand
uitvoerbestand = 'uitvoer.txt'  # Pad naar het uitvoerbestand
scheidingsteken = ' '          # Scheidingsteken: ' ' voor spatie, ',' voor komma, '\t' voor tab
# ─────────────────────────────────────────────────────────────────────────────
```

```python
# Bestand inlezen
df = pd.read_csv(invoerbestand, sep=scheidingsteken, header=None, engine='python')

print(f'Ingelezen: {len(df)} rijen, {len(df.columns)} kolommen')
print('Eerste rijen:')
df.head()
```

```python
# Enkel de eerste 6 kolommen bewaren
df_gefilterd = df.iloc[:, :6]

print(f'Na filteren: {len(df_gefilterd.columns)} kolommen')
df_gefilterd.head()
```

```python
# Resultaat wegschrijven
df_gefilterd.to_csv(uitvoerbestand, sep=scheidingsteken, header=False, index=False)

print(f'Opgeslagen als: {uitvoerbestand}')
```
